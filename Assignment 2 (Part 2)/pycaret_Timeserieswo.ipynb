# **Exploratory Data Analysis**

```python
df = pd.read_csv('/content/drive/MyDrive/Data_Mining_CMPE_255/time_series_wo.csv')
from pycaret.time_series import *
y = df['Value']
```

```python
fh = 30
fold = 4
fig_kwargs = {
    "renderer": "png",
    "width": 1000,
    "height": 600,
}
```

```python
Analysis = TSForecastingExperiment()
Analysis = setup(data=y, fh=fh, fold = fold, fig_kwargs=fig_kwargs)
Analysis.setup(data=y, fh=fh, fold = fold, fig_kwargs=fig_kwargs)
```

```python
Analysis.plot_model()
Analysis.plot_model(plot="periodogram")
Analysis.plot_model(plot="fft")
```

```python
Analysis.plot_model(plot="train_test_split")
Analysis.plot_model(plot="cv")
```

# **Statistical Tests**

```python
Analysis.check_stats(test="summary")
```

```python
Analysis.check_stats(test="stationarity")
```

```python
Analysis.check_stats(test="white_noise")
```

```python
Analysis.check_stats()
```

# **Modeling**

```python
Predictor = Analysis.create_model("knn_cds_dt")
```

```python
prediction = Analysis.predict_model(Predictor)
prediction
```

```python
Analysis.plot_model(estimator=Predictor)
```

## **Model Tuning**

```python
tuned_Predictor = Analysis.tune_model(Predictor)
```

```python
Analysis.plot_model(tuned_Predictor)
```

```python
Analysis.plot_model([Predictor, tuned_Predictor], data_kwargs={"labels": ["Baseline", "Tuned"]})
```

```python
final_Predictor = Analysis.finalize_model(tuned_Predictor)
Analysis.plot_model(final_Predictor)
```

## **Saving Model**

```python
save_model(final_Predictor, 'Time_Series_wo')
```
