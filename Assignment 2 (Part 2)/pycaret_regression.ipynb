# **Regression**

```python
df = p.read_csv('/content/drive/MyDrive/Data_Mining_CMPE_255/Crab_Age.csv')
from pycaret.regression import *
```

## **Setup**

```python
regress = RegressionExperiment()
regress = setup(df, target='Age')
regress.setup(df, target= 'Age')
```

## **Comparing Models**

```python
Age = regress.compare_models()
```

## **Analyzing the best Model**

```python
plot_model(Age, 'residuals')
```

```python
plot_model(Age, 'feature')
```

```python
plot_model(Age, 'error')
```

```python
evaluate_model(Age)
```

## **Prediction**

```python
Crab_age_predictor = predict_model(Age)
```

```python
save_model(Age, 'Crab_Age')
```
